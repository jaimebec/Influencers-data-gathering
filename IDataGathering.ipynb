```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup as soup
import pandas as pd
from IPython.core.display import HTML

"""
   list_of_dicts = [
{"name": "Ginger", "breed": "Dachshund", "height_cm": 22,
"weight_kg": 10, "date_of_birth": "2019-03-14"}, {"name": "Scout", "breed": "Dalmatian", "height_cm": 59,
"weight_kg": 25, "date_of_birth": "2019-05-09"} ]

"""

#findAll('table', {'class': 'items'})
#find('thead')
#find_all('th')

#users test: anniejaffrey, kellystrackofficial

TestList = ['anniejaffrey', 'kellystrackofficial', 'laurajadestone']
ListInfluencer = []

# aqui empieza el loop

counter = 0
print(len(ronda))
print(" "*(len(ronda)-1), "]", len(ronda))
for i in ronda:
    
    try:
        counter += 1
        print('X', end='')

        ColumnsNames = []

        my_url = 'https://socialblade.com/instagram/user/' + i

        req = Request(my_url, headers={'User-Agent': 'Mozilla/5.0'})

        web_byte = urlopen(req).read()

        webpage = web_byte.decode('utf-8')

        # html parsing
        page_soup = soup(webpage, 'html.parser')

        # dictionary to store the data of the influencer

        InfluencerDict = {}

        # digging the web into the data

        # Name located in a different format than the other data
        # Adding it to the dictionary

        filter1 = page_soup.find('div', {'id': 'YouTubeUserTopInfoBlockTop'})
        filter2 = filter1.find('h2').text

        InfluencerDict['Name'] = filter2

        # Looking for the rest of the data. Located in the same table

        # Columns names

        filter1 = page_soup.find('div', {'id': 'YouTubeUserTopInfoBlock'}).findAll('div', {'class': 'YouTubeUserTopInfo'})

        for i in range(len(filter1) - 1):
            ColumnsNames.append(filter1[i].find('span', {'class': 'YouTubeUserTopLight'}).text)

        # Filling the data

        filter1 = page_soup.find('div', {'id': 'YouTubeUserTopInfoBlock'}).findAll('div', {'class': 'YouTubeUserTopInfo'})

        x = 0
        for i in range(len(filter1) - 1):
            if i == 3:
                filter2 = filter1[3].find_all('span')
                InfluencerDict[ColumnsNames[x]] = filter2[2].text.strip()
                x += 1
                continue

            filter2 = filter1[i].find_all('span')
            InfluencerDict[ColumnsNames[x]] = filter2[1].text
            x += 1

        ListInfluencer.append((InfluencerDict))
    except AttributeError:
        continue

# Aca termina el primer loop
# ListInfluencer es la lista donde iran todos los dict para ser convertida en df


DfInfluencer = pd.DataFrame(ListInfluencer)

DfInfluencer
```

# Data Cleaning

```python
#Data Cleaning

# Getting the first name
# CleaningName[0] = First Name

CleaningName = DfInfluencer["Name"].str.split(" ", n = 1, expand = True)
CleaningName[0] = CleaningName[0].str.title()
CleaningName

#

# Getting the full name and account
# CleaningAcc[0] = full name
# CleaningAcc[1] = Account

CleaningAcc = DfInfluencer["Name"].str.split("@", n = 1, expand = True)
CleaningAcc[0] = CleaningAcc[0].str.title()
CleaningAcc[1] = ('@' + CleaningAcc[1])
CleaningAcc

#

#adding clean columns

DfInfluencer['Full Name'] = CleaningAcc[0]
DfInfluencer['Instagram Account'] = CleaningAcc[1]

# Cleaning numbers and changing type

DfInfluencer['Media Uploads'] = DfInfluencer['Media Uploads'].str.replace(',', '')
DfInfluencer['Engagement Rate'] = DfInfluencer['Engagement Rate'].str.replace('%', '')
DfInfluencer['Engagement Rate'] = DfInfluencer['Engagement Rate'].str.replace('--', '0')
DfInfluencer['Followers'] = DfInfluencer['Followers'].str.replace(',', '')
DfInfluencer['Following'] = DfInfluencer['Following'].str.replace(',', '')
DfInfluencer['AVG Likes'] = DfInfluencer['AVG Likes'].str.replace(',', '')
DfInfluencer['AVG Comments'] = DfInfluencer['AVG Comments'].str.replace(',', '')

DfInfluencer = DfInfluencer.astype({'Media Uploads': int, 'Engagement Rate': float, 'Followers': int, 'AVG Likes': float, 'AVG Comments':float, 'Following':int})

# Droping dirty column

# adding link page

DfInfluencer['IG Link'] = DfInfluencer['Instagram Account'].str.replace('@', 'https://www.instagram.com/')

DfInfluencer
```
